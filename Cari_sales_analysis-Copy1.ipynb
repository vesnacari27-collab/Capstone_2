```python
# Merge sales with product and category info
sales_full = (StoreSales_df
    .merge(StoreDetail_df[['Store ID', 'Territory Manager']], on='Store ID')
    .merge(Products_df[['Prod Num', 'CategoryID', 'SubcategoryID']], on='Prod Num')
    .merge(ProductCategories_df[['CategoryID', 'Category', 'SubcategoryID', 'Subcategory']],
           on=['CategoryID', 'SubcategoryID'])
)

# Filter to assigned territories
territory_cat = sales_full[
    sales_full['Territory Manager'].isin(['Len Jensen', 'See Ellefson'])
].copy()

# Convert to datetime FIRST, then extract YearMonth
territory_cat['Transaction Date'] = pd.to_datetime(territory_cat['Transaction Date'])
territory_cat['YearMonth'] = territory_cat['Transaction Date'].dt.to_period('M')
```

```python
# Monthly transaction counts by category and territory
txn_monthly = (territory_cat
    .groupby(['Territory Manager', 'YearMonth', 'Category'])
    .size().reset_index(name='Transactions'))

# Average monthly transactions per category (easier to compare across territories)
txn_avg = (txn_monthly
    .groupby(['Territory Manager', 'Category'])['Transactions']
    .mean().round(1).reset_index())

txn_pivot = txn_avg.pivot(index='Category', columns='Territory Manager', values='Transactions')
print("Average Monthly Transactions by Category")
print(txn_pivot.sort_values('See Ellefson', ascending=False))
```

```python
# Monthly revenue by category and territory
rev_monthly = (territory_cat
    .groupby(['Territory Manager', 'YearMonth', 'Category'])['Sale Amount']
    .sum().reset_index())

rev_avg = (rev_monthly
    .groupby(['Territory Manager', 'Category'])['Sale Amount']
    .mean().round(2).reset_index())

rev_pivot = rev_avg.pivot(index='Category', columns='Territory Manager', values='Sale Amount')
print("Average Monthly Revenue by Category")
print(rev_pivot.sort_values('See Ellefson', ascending=False).map('${:,.2f}'.format))
```

```python
# Average sale value per transaction — reveals which categories drive big-ticket purchases
avg_txn = (territory_cat
    .groupby(['Territory Manager', 'Category'])
    .agg(Avg_Sale=('Sale Amount', 'mean'), Total_Txns=('Sale Amount', 'count'))
    .reset_index())

print("Average Transaction Value by Category")
print(avg_txn.sort_values(['Territory Manager', 'Avg_Sale'], ascending=[True, False]).to_string(index=False))
```

```python
# Compare 2022-2023 vs 2024-2025 to identify growing categories
territory_cat['Half'] = territory_cat['Transaction Date'].dt.year.apply(
    lambda x: '2022-2023' if x <= 2023 else '2024-2025'
)

trend = (territory_cat
    .groupby(['Territory Manager', 'Half', 'Category'])['Sale Amount']
    .sum().reset_index())

trend_pivot = trend.pivot_table(
    index=['Territory Manager', 'Category'], columns='Half', values='Sale Amount'
)
trend_pivot['Growth %'] = (
    (trend_pivot['2024-2025'] - trend_pivot['2022-2023']) / trend_pivot['2022-2023'] * 100
).round(1)

print("Revenue Growth by Category: 2022-2023 vs 2024-2025")
print(trend_pivot.to_string())
```

### Findings: Transactions & Revenue by Product Category

Six product categories are sold across both territories (South Carolina & New York). The table below shows average monthly transactions & revenue per category.

**Category overview:**

| Category | Average Monthly Txns (SC) | Average Monthly Txns (NY) | Average Monthly Revenue (SC) | Average Monthly Revenue (NY) |
|---|---|---|---|---|
| Technology & Accessories | 20.4 | 134.4 | $9,836 | $63,982 |
| Stationery & Supplies | 18.9 | 136.4 | $190 | $1,371 |
| Apparel & Merchandise | 18.9 | 127.7 | $634 | $4,129 |
| Art Supplies | 18.5 | 122.5 | $583 | $3,750 |
| Textbooks | 12.2 | 92.2 | $2,151 | $15,977 |

**Key observations:**

**Technology & Accessories** is by far the top revenue earner in both territories generating an average of 9,836 dollars per month in SC and 63,982 dollars per month in New York despite having a similar transaction count to Stationary & Apparel. This is explained by the average transaction value: Technology items aeverage 483 dollars per sale compared to just 10 dollars for Stationary and 34 dollars for Apparel. High-value items like laptops, tablets, & headphones secure overall revenue in both territories. 

**Textbooks** rank second in revenue with 2,151 dollars per month in SC and 15,977 dollars per month in NY. But they also rank fifth in transaction volume due to high average sale value (around 175 dollars). This category likely shows seasonal spikes around August-September because of back to school and in January because of the spring semester. 

**Stationary & Supplies** is the highest category in volume by transactions but generates the least in revenue per sale (10 dollars average) making it a source of traffic rather than revenue. 



### Findings: Category Growth Trends 

Comparing the first 2 years of data (2022-2023) to the most recent 2 years (2024-2025) reveals meaningful growth
patterns:

| Category | SC Growth | NY Growth |
|---|---|---|
| Art Supplies | +90.4% | +95.0% |
| Books (General) | +78.7% | +92.1% |
| Stationery and Supplies | +84.9% | +86.9% |
| Apparel and Merchandise | +75.5% | +76.0% |
| Technology & Accessories | +66.1% | +56.7% |
| Textbooks | +10.0% | +11.8% |

**Key observations:**

**Art Supplies, Books, & Stationery** are the fastest-growing categories in both 
territories — all growing 75–95% from the first half to the second half of the data 
period which therfore suggests rising customer interest in a range of creative & educational products.

**Technology & Accessories** is growing more slowly relative to other categories 
(~57–66%), which is notable given it still dominates revenue. This may reflect market 
saturation or increased price competition.

**Textbooks** show the least growth (~10–12%) — likely because this is a needs-based 
purchase category with limited room to expand beyond the existing student customer base.

## Marketing Recommendation: 

Based on the full analysis of both sales territories (monthly revenue, store rankings, top customers, & category trends) the following recommendations are made for where to focus marketing attention in the next quarter. 

#### 1. Prioritize Art Supplies & Books in both territories 
Art Supplies & Books are currently the fastest-growing categories in both territories. Art Supplies, in particular, is seeing very strong growth despite still generating relatively low monthly revenue compared to larger categories. This suggests there is significant room for expansion. Targeted promotions — such as "Buy-one-get-one" discounts on art supplies and books can help build on this momentum while attracting more creative-based customers who wouldn't engage with tech-centered promotions as strongly. 

#### 2. Target Back-to-School Season for Textbooks 
In both territories the textbooks rank second in revenue but show the lowest growth because because it's a time sensitive and needs-based category. It could increase revenue by including more seasonal promotions for students and school staff which would occur around July-August and again in January. Bundling textbook purchases with another category such as Stationary & Supplies would increase average sales with targeted campaigns.  

#### 3. Rewards Program Engagement in South Carolina stores
Since the South Carolina territory has 2 stores and their revenue is much different than the NY stores they still show top customers spending as much as the ones in the NY stores. Despite only 10% of transactions being linked to a rewards ID they can still work with customers to have a better engagement with them trhough in-store rewards enrollment and outreach. Working on increasing the enrollment with the customer base would help significantly with better marketing towards reward members and customer data overall.  

## Monthly Total Revenue by Territory 

```python
# Monthly Total Revenue by Territory — Line Chart

# Merge sales with store detail to get territory manager
monthly_revenue = (StoreSales_df
    .merge(StoreDetail_df[['Store ID', 'Territory Manager']], on='Store ID')
)

# Filter to assigned territories
monthly_revenue = monthly_revenue[
    monthly_revenue['Territory Manager'].isin(['Len Jensen', 'See Ellefson'])
].copy()

# Convert Transaction Date and extract YearMonth
monthly_revenue['Transaction Date'] = pd.to_datetime(monthly_revenue['Transaction Date'])
monthly_revenue['YearMonth'] = monthly_revenue['Transaction Date'].dt.to_period('M')

# Group by territory and month
monthly_rev_grouped = (monthly_revenue
    .groupby(['Territory Manager', 'YearMonth'])['Sale Amount']
    .sum()
    .reset_index()
)
monthly_rev_grouped['YearMonth'] = monthly_rev_grouped['YearMonth'].dt.to_timestamp()

# Split by territory
sc = monthly_rev_grouped[monthly_rev_grouped['Territory Manager'] == 'Len Jensen']
ny = monthly_rev_grouped[monthly_rev_grouped['Territory Manager'] == 'See Ellefson']

# Plot
fig, ax = plt.subplots(figsize=(14, 5))

ax.plot(sc['YearMonth'], sc['Sale Amount'], label='Len Jensen (South Carolina)', color='steelblue', linewidth=2)
ax.plot(ny['YearMonth'], ny['Sale Amount'], label='See Ellefson (New York)', color='darkorange', linewidth=2)

ax.set_title('Monthly Total Revenue by Territory (Jan 2022 – Dec 2025)', fontsize=14)
ax.set_xlabel('Month')
ax.set_ylabel('Total Revenue ($)')
ax.legend()
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Average Monthly Revenue by Product Category — Grouped Bar Chart

# Merge sales with store, product, and category info
cat_revenue = (StoreSales_df
    .merge(StoreDetail_df[['Store ID', 'Territory Manager']], on='Store ID')
    .merge(Products_df[['Prod Num', 'CategoryID', 'SubcategoryID']], on='Prod Num')
    .merge(ProductCategories_df[['CategoryID', 'SubcategoryID', 'Category']],
           on=['CategoryID', 'SubcategoryID'])
)

# Filter to assigned territories
cat_revenue = cat_revenue[
    cat_revenue['Territory Manager'].isin(['Len Jensen', 'See Ellefson'])
].copy()

cat_revenue['Transaction Date'] = pd.to_datetime(cat_revenue['Transaction Date'])
cat_revenue['YearMonth'] = cat_revenue['Transaction Date'].dt.to_period('M')

# Average monthly revenue per category per territory
cat_grouped = (cat_revenue
    .groupby(['Territory Manager', 'YearMonth', 'Category'])['Sale Amount']
    .sum()
    .reset_index()
    .groupby(['Territory Manager', 'Category'])['Sale Amount']
    .mean()
    .reset_index()
)

# Pivot for side-by-side bars
cat_pivot = cat_grouped.pivot(index='Category', columns='Territory Manager', values='Sale Amount')
cat_pivot = cat_pivot.sort_values('See Ellefson', ascending=False)

# Plot
fig, ax = plt.subplots(figsize=(12, 6))

cat_pivot.plot(kind='bar', ax=ax, color=['steelblue', 'darkorange'], width=0.7)

ax.set_title('Average Monthly Revenue by Product Category', fontsize=14)
ax.set_xlabel('Product Category')
ax.set_ylabel('Avg Monthly Revenue ($)')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.legend(title='Territory Manager')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```
